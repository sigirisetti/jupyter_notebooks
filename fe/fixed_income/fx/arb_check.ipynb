# Arb Check aganist Reference Price

```python
import pandas as pd
import numpy as np

df = pd.read_csv("price_collar_usage.csv")
df
```

```python
tolerance = 0.01
```

```python
df["minRefPrice"] = df["refPrice"]*(1 - 0.5 * tolerance)
df["maxRefPrice"] = df["refPrice"]*(1 + 0.5 * tolerance)
df
```

```python
df["tobBidArbable"] = df["bid"] - df["maxRefPrice"] > 1e-8
df["tobAskArbable"] = df["minRefPrice"] - df["ask"] > 1e-8
df
```
