# Price Collar Usage

```python
import pandas as pd
import numpy as np

df = pd.read_csv("price_collar_usage.csv")
df
```

## Add min bid / max ask using ref price

Set a percentage of ref price to be used to generate collar

```python
percent_use = 0.5
```

```python
df["bidMin"] = df["refPrice"]*(1-percent_use)
df["askMax"] = df["refPrice"]*(1+percent_use)
df
```

```python
df["bidPCUsage"] = 100 * np.minimum(1, np.maximum(0, (df["refPrice"]-df["bid"])/(df["refPrice"]-df["bidMin"])))
df["askPCUsage"] = 100 * np.minimum(1, np.maximum(0, (df["ask"]-df["refPrice"])/(df["askMax"]-df["refPrice"])))
#df["bidPricesRatio"] = (df["refPrice"]-df["bid"])/(df["refPrice"]-df["bidMin"])
#df["askPricesRatio"] = (df["ask"]-df["refPrice"])/(df["askMax"]-df["refPrice"])
df
```
